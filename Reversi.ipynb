```
def evaluacion(state, player):
    """A very complex evaluation function"""
    othello = Othello()
    state_legal_moves = othello.legal_moves(state.moves, state.board, state.to_move)

    black_tiles = [loc for loc, tile in state.board.items() if tile == 'B']

    # Feature 1: Difference between black and white
    feature_one = 0
    for disc in state.board.values():
        feature_one += 1 if disc == 'B' else -1

    # Feature 2: Power spots
    feature_two = 0
    power_spots_1 = [(1,1), (1,8), (8,1), (8,8)]
    power_spots_2 = [(1,3), (1,6), (3,1), (3,8), (6,1), (6,8), (8,3), (8,6)]
    power_spots_3 = [(3,3), (3,6), (6,3), (6,6)]

    for spot in power_spots_1:
        if spot in black_tiles:
            feature_two += 10

    for spot in power_spots_2:
        if spot in black_tiles:
            feature_two += 7

    for spot in power_spots_3:
        if spot in black_tiles:
            feature_two += 4

    # Feature 3: Try to have more possible movements
    feature_three = len(state_legal_moves)
    
    evaluation = feature_one + feature_two + feature_three
    if player == 'B':
        return evaluation
    else:
        return -evaluation

wins = 0
for _ in range(100):
    game = play_game(Othello(), alphabeta_player(eval_fn=evaluacion), random_player, show=False)
    wins += 1 if game > 0 else 0
print(wins)

# game = play_game(Othello(), alphabeta_player(eval_fn=evaluacion), alphabeta_player(), show=False)
```
